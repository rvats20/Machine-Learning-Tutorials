**SPLIT DATASET INTO TRAINING & VALIDATION**

```python

def split_dataset(dataset, validation_split=0.2, shuffle=True, seed=None):
    # Get the total number of samples
    num_samples = len(dataset)

    # Calculate the number of samples for validation
    num_validation_samples = int(validation_split * num_samples)

    # Optionally shuffle the dataset
    if shuffle:
        dataset = dataset.shuffle(buffer_size=num_samples, seed=seed)

    # Split the dataset into training and validation sets
    train_dataset = dataset.skip(num_validation_samples)
    validation_dataset = dataset.take(num_validation_samples)

    return train_dataset, validation_dataset
```

```python
# Split the dataset
train_dataset, validation_dataset = split_dataset(full_dataset, validation_split=0.2, shuffle=True, seed=123)
```

**LENET CNN MODEL**

```python
IM_SIZE = 224
num_classes = 5

lenet_model = tf.keras.Sequential([
    InputLayer(input_shape=(IM_SIZE, IM_SIZE, 3)),
    Rescaling(1.0 / 127.5, offset=-1),

    Conv2D(filters=6, kernel_size=5, strides=1, padding='same', activation='relu'),
    BatchNormalization(),
    MaxPool2D(pool_size=2, strides=2),

    Conv2D(filters=16, kernel_size=5, strides=1, padding='same', activation='relu'),
    BatchNormalization(),
    MaxPool2D(pool_size=2, strides=2),

    Conv2D(filters=120, kernel_size=5, strides=1, padding='same', activation='relu'),
    BatchNormalization(),
    MaxPool2D(pool_size=2, strides=2),

    Conv2D(filters=120, kernel_size=5, strides=1, padding='same', activation='relu'),
    BatchNormalization(),
    MaxPool2D(pool_size=2, strides=2),

    Conv2D(filters=84, kernel_size=5, strides=1, padding='same', activation='relu'),
    BatchNormalization(),
    MaxPool2D(pool_size=2, strides=2),

    Conv2D(filters=84, kernel_size=5, strides=1, padding='same', activation='relu'),
    BatchNormalization(),
    MaxPool2D(pool_size=2, strides=2),

    Flatten(),
    Dense(100, activation='relu'),
    BatchNormalization(),
    Dense(num_classes, activation='softmax')
])

lenet_model.summary()
```

```python
y_true = [0,1,0,0]
y_pred = [0.6,0.51,0.94,1]
bce = tf.keras.losses.BinaryCrossentropy()
bce(y_true, y_pred)
```

**LEARNING RATE & ADAM OPTIMIIZATION**

```python
learning_rate = 0.001
optimizer = tf.keras.optimizers.Adam(learning_rate=learning_rate)

lenet_model.compile(optimizer=optimizer,
                    loss='categorical_crossentropy',
                    metrics=['accuracy'])
```

**LOAD IMAGE**

```python
# Function to load and prepare an image
def load_image(img_path):
    img = keras_image.load_img(img_path, target_size=(224, 224))  # Ensure the image size matches the model input size
    img_array = keras_image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)
    return img_array
```

**APPLY PREDICTION**

```python
# Function to make predictions on a single image
def predict_image(img_path):
    image = load_image(img_path)
    predictions = lenet_model.predict(image)
    predicted_class_index = np.argmax(predictions)
    predicted_class_name = class_names[predicted_class_index]
    return predicted_class_name
```

**CONFUSION MATRIX**

```python
true_labels = []
predicted_labels = []

for images, labels in test_dataset:
    # Make predictions using your model
    predictions = lenet_model.predict(images)
    # Convert one-hot encoded labels to class labels
    true_labels.extend(np.argmax(labels.numpy(), axis=1))
    predicted_labels.extend(np.argmax(predictions, axis=1))
```

```python
labels = ['DR', 'Mild DR ', 'moderate DR', 'severe DR', 'proliferative DR']


conf_matrix = confusion_matrix(true_labels, predicted_labels)


plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, cmap='Blues', fmt='d', xticklabels=labels, yticklabels=labels)
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix')
plt.show()
```

**ROC CURVE GRAPH**

```python
true_labels = []
predicted_probabilities = []

# Iterate over the test dataset to obtain true labels and predicted probabilities
for images, labels in test_dataset:
    # Make predictions using your model
    predictions = lenet_model.predict(images)
    # Append true labels and predicted probabilities
    true_labels.extend(np.argmax(labels.numpy(), axis=1))
    predicted_probabilities.extend(predictions)

# Convert lists to numpy arrays
true_labels = np.array(true_labels)
predicted_probabilities = np.array(predicted_probabilities)
```

```python
fpr = dict()
tpr = dict()
roc_auc = dict()
for i in range(num_classes):
    fpr[i], tpr[i], _ = roc_curve(true_labels == i, predicted_probabilities[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])
```

```python
plt.figure(figsize=(10, 8))
for i in range(num_classes):
    plt.plot(fpr[i], tpr[i], label='ROC curve (area = {:.2f}) for class {}'.format(roc_auc[i], i))

plt.plot([0, 1], [0, 1], 'k--')  # Diagonal line
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc="lower right")
plt.show()
```
